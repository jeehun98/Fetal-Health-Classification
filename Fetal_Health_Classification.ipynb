데이터 분리

```python
target_data = data[['fetal_health']]

feature_data = data.drop(['fetal_health'], axis=1, inplace=False)

print(len(data.columns), len(feature_data.columns), len(target_data.columns))
```

```python
corr_matrix = feature_data.corr()
```

특성 사이의 상관관계를 확인하기 위해

```python
corr_matrix
```

상관관계 조사, 특성 간의 표준 상관 계수 standard correlation coefficient( 피어슨의 r) 를 corr() 메서드를 이용해 쉽게 계산할 수 있다.

```python
from pandas.plotting import scatter_matrix

attributes = ["baseline value", "histogram_mode", "histogram_mean", "histogram_median"]
scatter_matrix(feature_data[attributes], figsize=(12, 8))
```

테스트 세트 분리

```python
from sklearn.model_selection import train_test_split

feature_train, feature_test, target_train, target_test = train_test_split(feature_data, target_data, test_size = 0.2, random_state=42, stratify=target_data)
```

특성 스케일링 파이프 라인

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
  ('std_scaler', StandardScaler())
])

feature_train_scaled = num_pipeline.fit_transform(feature_train)
feature_test_scaled = num_pipeline.transform(feature_test)
```

```python
feature_test_scaled
```

모델 선택과 훈련

서포트 벡터 머신

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(feature_train_scaled, target_train)
svm_clf.score(feature_train_scaled, target_train)
```

확률적 경사 하강법 SGD

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier()
sgd_clf.fit(feature_train_scaled, target_train)
sgd_clf.score(feature_train_scaled, target_train)
```

결정 트리

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.fit(feature_train_scaled, target_train)
tree_clf.score(feature_train_scaled, target_train)
```

결정 트리 특성별 중요도 확인

```python
tree_clf.feature_importances_
```

교차 검증을 통한 확인, 훈련, 테스트 세트

```python
from sklearn.model_selection import cross_val_score

cross_val_score(svm_clf, feature_train_scaled, target_train, cv=3, scoring="accuracy")
```

교차 검증 함수 생성, 테스트 점수와 같이 확인

```python
def cross_val (model, feature, target, cv):
  return cross_val_score(model, feature, target, cv=cv, scoring="accuracy")
```

```python
svm_cross_score = cross_val(svm_clf, feature_train_scaled, target_train, 3)
svm_test_score = svm_clf.score(feature_test_scaled, target_test)

print(svm_cross_score, svm_test_score)
```

```python
sgd_cross_score = cross_val(sgd_clf, feature_train_scaled, target_train, 3)
sgd_test_score = sgd_clf.score(feature_test_scaled, target_test)

print(sgd_cross_score, sgd_test_score)
```

```python
dt_cross_score = cross_val(tree_clf, feature_train_scaled, target_train, 3)
dt_test_score = tree_clf.score(feature_test_scaled, target_test)

print(dt_cross_score, dt_test_score)
```
